```python
import numpy as np
import pandas as pd
import seaborn as sns

def get_file_using_PublicURL(public_url):
  file_id_url = public_url.split('/')[-2]
  return 'https://drive.google.com/uc?id=' + file_id_url

dataset = pd.read_csv( get_file_using_PublicURL('https://drive.google.com/file/d/18LJnC1W2HUaHFD7WCP8tOm56Kk9FwMdS/view?usp=drive_link'), index_col = 0 )
churn_telco = pd.read_csv( get_file_using_PublicURL('https://drive.google.com/file/d/1FYHXc-jV5Jl6s2ee6voRc4i_ouMcsxtX/view?usp=drive_link'), index_col = 0 )
first_telco = pd.read_csv( get_file_using_PublicURL('https://drive.google.com/file/d/1CCY55fjvKFSEe_6aHrm0e_DDJftrXmPa/view?usp=drive_link'), index_col = 0 )
```

```python
dataset.info()
print('\nCount NA rows\n',dataset.isna().sum())
print('\nCount duplicated rows\n',dataset.duplicated().sum())
# dataset[(dataset.duplicated())]
dataset.drop_duplicates(inplace=True)
print('\nAfter removing the duplicated rows\n')
dataset.info()
```
